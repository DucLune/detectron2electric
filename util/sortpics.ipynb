```python
# 从文件中获取要拷贝的文件的信息
def shuffle_list(listfile):
    shuffled_lists = []
    with open(listfile, 'r', encoding='utf-8') as f:
        lists = f.readlines()
        random.shuffle(lists)
        for list in lists:
            shuffled_lists.append(str(list))
    return shuffled_lists
shuffle_list("./val.txt")
```

```python
def split_train_test(listfile="./val.txt", portion=0.7):
    shuffled_list = shuffle_list(listfile)
    traintxt = open("train.txt", "w+")
    testtxt = open("test.txt", "w+")
    for i in range(len(shuffled_list)):
        if i <= 0.7*len(shuffled_list):
            traintxt.write(shuffled_list[i])
        else:
            testtxt.write(shuffled_list[i])


split_train_test()
```

```python

# python读取文件夹下所有文件名并写入.txt文件
file_list = os.listdir(os.path.join(src_dirname,"Annotations"))
random.shuffle(file_list)
textfile = open(os.path.join(dest_dirname,"ImageSets","Main","test.txt"), "w+")
for file in file_list:
    title = file.split(".")
    if(title[0][-1] != '5'):
        textfile.write(title[0]+'\n')
textfile.close()
```

```python
# 从文件中获取要拷贝的文件的信息
def get_filename_from_txt(file):
    filename_lists = []
    with open(file, 'r', encoding='utf-8') as f:
        lists = f.readlines()
        for list in lists:
            filename_lists.append(str(list).strip('\n')+'.jpg')
    return filename_lists
```

```python
# 拷贝文件到新的文件夹中
def mycopy(srcpath, dstpath, filename):
    if not os.path.exists(srcpath):
        print("srcpath not exist!")
    if not os.path.exists(dstpath):
        print("dstpath not exist!")
    for root, dirs, files in os.walk(srcpath, True):

        if filename in files:
            # 如果存在就拷贝
            shutil.copy(os.path.join(root, filename), dstpath)
        else:
            # 不存在的话将文件信息打印出来
            print(filename)
```

```python
if __name__ == "__main__":
    # 执行获取文件信息的程序
    filename_lists = get_filename_from_txt(file_path)
    # 根据获取的信息进行遍历输出
    for filename in filename_lists:
        mycopy(oldpath, newpath, filename)
```
